## Parte 4

```python
for i,df in enumerate ([df,df1,df2,jsdf,jsdf2,jsdf3]):
    df.to_csv(f"./dataing/{i}.csv")
    df.to_excel(f"./dataing/{i}.xlsx")
    df.to_json(f"./dataing/{i}.json")
```

```python
for i,df in enumerate ([df,df1,df2,jsdf,jsdf2,jsdf3]):
    with open(f"./dataing/{i}.pickle", 'wb') as f:
        pickle.dump(df,f,protocol=4)
```
